### Formula for Min-Max Scaling

Given a feature $\ X $ with a minimum value of $ \text{min}_X $ and a maximum value of $ \text{max}_X $, the scaling to the $[0, 1]$ range is done using the following formula:

$$ \ X_{\text{scaled}} = \frac{X - \text{min}_X}{\text{max}_X - \text{min}_X} \ $$

#### Explanation

- **$\ X \ $**: The original data value.
- **$ \text{min}_X $**: The minimum value of the feature in the dataset.
- **$\ \text{max}_X \$**: The maximum value of the feature in the dataset.
- **\$ X_{\text{scaled}} \$**: The scaled value of \$ X \$ that will be in the range \$ [0, 1]\$.
